```python
import pandas as pd
```

```python
amazon_data = pd.read_csv('amazon_Reviews.csv')
amazon_data.head(1)
```

```python
df = amazon_data
```

```python
df = df.rename(columns={"reviews.date":"reviews_date",
                                "reviews.rating": "reviews_rating_score",
                                "reviews.text":"reviews_text","reviews.title":"reviews_title",
                        "reviews.numHelpful":"reviews_numHelpful",
                       "reviews.doRecommend":"reviews_doRecommend"})
```

```python
# Into CSV
df.to_csv('amazon_data_clean.csv')
```

```python
df_small = df.copy()
```

```python
df_small = df_small[["brand","categories","manufacturer","reviews_date","reviews_doRecommend","reviews_rating_score","reviews_text","reviews_title","review_rating"]]
```

```python
df_small.columns
```

```python
# Into CSV
df_small.to_csv('amazon_data_small_clean.csv')
```
